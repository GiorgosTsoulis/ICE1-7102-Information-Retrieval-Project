## Βήμα 2: Προεπεξεργασία κειμένου (Text Processing)
Στο  βήμα αυτό θα προχωρήσουμε σε προεπεξεργασία του κειμενικού περιεχόμενου κάθε άρθρου του συνόλου δεδομένων που επιλέξαμε εκτελώντας tokenization, lemmatization, stop-word removal και special characters removal

Αρχικά εισάγουμε στο πρόγραμμα μας τις απαραίτητες βιβλιοθήκες και φορτώνουμε το σύνολο δεδομένων σε ένα Dataframe του pandas προκειμένου να μπορούμε να το επεξεργαστούμε.

```python
# Εισαγωγή βιβλιοθηκών
import pandas as pd
import json
import nltk
from nltk.tokenize import word_tokenize
import string

# Φόρτωση του dataset σε DataFrame
data = pd.read_json(r"C:\Users\07dig\OneDrive\articles.json")
```

### Tokenization
Αρχικά κάνουμε tokenization στο κειμενικό περιεχόμενο κάθε άρθρου, προκειμένου να το μετατρέψουμε σε λέξεις(tokens).

```python
# Ορισμός συνάρτησης tokenization που χρησιμοποιεί τη word_tokenize
def tokenize(text):
    # Μετατροπή του κειμένου σε πεζά
    text = text.lower()
    # Επιστροφή του tokenized κειμένου
    return word_tokenize(text)

# Εφαρμογή της συνάρτησης tokenize στη στήλη "Body"
data['Body_tokens'] = data['Body'].apply(tokenize)

# Εμφάνιση του tokenized αποτελέσματος
print(data[['Title', 'Body_tokens']].head(10))
```

### Lemmatization
Στη συνέχεια κάνουμε lemmatization στις λέξεις(tokens) κάθε άρθρου προκειμένου να επιστραφούν στην αρχική, λεξικολογική τους μορφή.

```python
# Λήψη του λεξικού WordNet από τη βιβλιοθήκη NLTK
nltk.download('wordnet')

# Δημιουργία αντικειμένου WordNetLemmatizer για τη λημματοποίηση
wnl = nltk.WordNetLemmatizer()

# Ορισμός συνάρτησης λημματοποίηση
def lemmatization(tokens):
    # Εφαρμογή λημματοποίηση σε κάθε token
    return [wnl.lemmatize(t) for t in tokens]

# Εφαρμογή της συνάρτησης λημματοποίηση στα tokens της στήλης 'Body_tokens'
data['Body_lemmas'] = data['Body_tokens'].apply(lemmatization)

# Εμφάνιση των πρώτων 10 αποτελεσμάτων για τους τίτλους και τα λημματοποίηση σώματα
print(data[['Title', 'Body_lemmas']].head(10))
```

### Stop-word Removal
Έπειτα επιχειρούμε να αφαιρέσουμε τα stop-words, δηλαδή τις λέξεις που δεν προσφέρουν σημαντική πληροφορία, απο το περιεχόμενο κάθε άρθρου.

```python
# Λήψη της λίστας των stopwords από τη βιβλιοθήκη NLTK
nltk.download('stopwords')

# Δημιουργία ενός συνόλου (set) με τα stopwords της αγγλικής γλώσσας
stop_words = set(nltk.corpus.stopwords.words('english'))

# Ορισμός συνάρτησης για την αφαίρεση των stopwords
def remove_stopwords(tokens):
    # Επιστροφή μόνο των tokens που δεν ανήκουν στα stopwords
    return [t for t in tokens if t not in stop_words]

# Εφαρμογή της συνάρτησης αφαίρεσης stopwords στα λεμματοποιημένα tokens της στήλης 'Body_lemmas'
data['Body_no_stopwords'] = data['Body_lemmas'].apply(remove_stopwords)

# Εμφάνιση των πρώτων 10 αποτελεσμάτων για τους τίτλους και τα σώματα χωρίς stopwords
print(data[['Title', 'Body_no_stopwords']].head(10))
```

### Special Characters Removal
Τέλος για τον καθαρισμό του συνόλου δεδομένων που επιλέξαμε, θα απομακρύνουμε τους ειδκούς χαρακτήρες, που μπορύν να επηρεάσουν την ανάλυση του κειμένου κάθε άρθρου.

```python
# Ορισμός συνάρτησης για την αφαίρεση ειδικών χαρακτήρων
def remove_special_characters(tokens):
    # Διαγραφή των tokens που περιέχουν ειδικούς χαρακτήρες (π.χ. σημεία στίξης)
    cleaned_tokens = [token for token in tokens if token not in string.punctuation] 
    return cleaned_tokens        

# Εφαρμογή της συνάρτησης στη στήλη 'Body_no_stopwords' για την αφαίρεση ειδικών χαρακτήρων
data['Body_no_special_characters'] = data['Body_no_stopwords'].apply(remove_special_characters)    

# Εμφάνιση των πρώτων 10 αποτελεσμάτων για τους τίτλους και τα σώματα χωρίς ειδικούς χαρακτήρες
print(data[['Title', 'Body_no_special_characters']].head(10))
```

Στο τέλος των εργασίων που εκτελέσαμε για τον καθαρισμό του συνόλου δεδομένων, αποθηκεύουμε το τελικό επεξεργασμένο κείμενου κάθε άρθου σε μια νέα στήλη "Body_cleaned".

```python
# Αντιγραφή της στήλης 'Body_no_special_characters' στη νέα στήλη 'Body_cleaned'
data['Body_cleaned'] = data['Body_no_special_characters']

# Εμφάνιση των πρώτων 10 αποτελεσμάτων για τους τίτλους και τα καθαρισμένα σώματα κειμένου
print(data[['Title', 'Body_cleaned']].head(10))
```

Στη συνέχεια δημιουργούμε ένα νέο DataFrame με όνομα structured_output που περιλαμβάνει μόνο τις στήλες "ID", "Title" και "Body_cleaned". Τέλος αποθηκεύουμε το λεξικό structured_output στο παρακάτω path.

```python
# Δημιουργία ενός νέου DataFrame με τη ζητούμενη δομή
structured_output = data[['ID', 'Title', 'Body_cleaned']].to_dict(orient='records')

# Αποθήκευση του δομημένου αποτελέσματος σε αρχείο JSON
with open(r"C:\Users\07dig\OneDrive\structured_tokenized_articles.json", 'w') as f:
    json.dump(structured_output, f, indent=4)
```

## Βήμα 3: Ευρετήριο (Indexing)
Σε αυτό το βήμα, θα δημιουργήσουμε μια ανεστραμμένη δομή ευρετηρίου (inverted index) για να αντιστοιχίσουμε αποτελεσματικά τους όρους με τα έγγραφα στα οποία εμφανίζονται. Παράλληλα, θα υλοποιήσουμε μια δομή δεδομένων που θα αποθηκεύει το ευρετήριο αυτό.

Αρχικά κατασκευάζουμε το αντεστραμμένο ευρετήριο βάση των επεξεργασμένων δεδομένων που συλλέχθηκαν απο το Βήμα 2.

```python
import json
import pandas as pd

# Φόρτωση του dataset σε DataFrame
data = pd.read_json(r"C:\Users\07dig\OneDrive\structured_tokenized_articles.json")

# Δημιουργία αντεστραμμένου ευρετηρίου
inverted_index = {}

# Για κάθε γραμμή στο DataFrame
for i, doc in data.iterrows():  # Η iterrows() επιτρέπει πρόσβαση σε κάθε γραμμή
    doc_id = doc['ID']  # Το ID του εγγράφου
    tokens = doc['Body_cleaned']  # Μοναδικοί όροι του εγγράφου
    
    # Για κάθε όρο στο έγγραφο
    for token in tokens:  # Προσθήκη όρου στο ευρετήριο
        if token not in inverted_index: # Αν ο όρος δεν υπάρχει στο ευρετήριο
            inverted_index[token] = [] # Δημιουργία νέας λίστας για τον όρο
        inverted_index[token].append(doc_id) # Προσθήκη του ID του εγγράφου στη λίστα του όρου 

        if token in inverted_index:
            inverted_index[token] = list(set(inverted_index[token])) # Αφαίρεση διπλότυπων τιμών από τη λίστα του όρου
```

Στη συνέχεια το αντεστραμμένο ευρετήριο αποθηκεύεται σε αρχείο JSON και κατόπιν εκτυπώνεται για τους πρώτους 10 όρους του.

```python
# Αποθήκευση ευρετηρίου σε JSON
with open(r"C:\Users\07dig\OneDrive\inverted_index.json", 'w') as f:
    json.dump(inverted_index, f, indent=4)

# Εμφάνιση δείγματος του ευρετηρίου
for term, docs in list(inverted_index.items())[:10]:  # Εμφάνιση των πρώτων 10 όρων
    print(f"Term: {term}, Documents: {docs}")
```

## Βήμα 4: Μηχανή αναζήτησης (Search Engine)
Στο παρόν βήμα θα υλοποιήθεί μια μηχανή αναζήτησης με σκοπό την αναζήτηση ακαδημαϊκών εργασιών χρησιμοποιώντας την Python.

### Ερώτημα α: Επεξεργασία ερωτήματος (Query Processing)
Στο συγκεκριμένο ερώτημα θα αναπτύξουμε ένα module επεξεργασίας ερωτημάτων που θα προεπεξεργάζεται τα ερωτήματα που λαμβάνει από τον χρήστη, θα τα αναλύει και θα ανακτά τα σχετικά άρθρα χρησιμοπιώντας το αντεστραμένο ευρετήριο.

Αρχικά φορτώνουμε το αντεστραμμένο ευρετήριο από το path στο οποίο το είχαμε αποθηκεύσει στο Βήμα 3.

```python
import json

# Φόρτωση του αντεστραμμένου ευρετηρίου ως λεξικό
with open(r"C:\Users\07dig\OneDrive\inverted_index.json", 'r') as f:
    inverted_index = json.load(f)
```

Στη συνέχεια δημιουργούμε την συνάρτηση boolean_search με την οποία θα επιτύχουμε την αναζήτηση των όρων του ερωτήματος στο αντεστραμμένο ευρετήριο με χρήση των Boolean λειτουργιών AND, OR και NOT.

```python
# Συνάρτηση για την εκτέλεση Boolean αναζήτησης
def boolean_search(query, index):
    query_terms = query.lower().split()  # Χωρισμός ερωτήματος σε λέξεις και μετατροπή σε πεζά
    result_docs = None  # Αρχικοποίηση με None
    all_docs = set()  # Σύνολο με όλα τα έγγραφα στο ευρετήριο
    operation = None # Αρχικά χωρίς λογική λειτουργία
    
     # Δημιουργία του συνόλου όλων των εγγράφων
    for docs in index.values():
        all_docs.update(docs)
    
    for term in query_terms:  # Επεξεργασία κάθε όρου στο ερώτημα
        if term in ["and", "AND"]: # Εάν ο όρος είναι and ή AND...
            operation = "AND" # Ορίζεται η λογική λειτουργία ως AND
        elif term in ["or", "OR"]: # Εάν ο όρος είναι or ή OR...
            operation = "OR" # Ορίζεται η λογική λειτουργία ως OR
        elif term in ["not", "NOT"]: # Εάν ο όρος είναι not ή NOT...
            operation = "NOT" # Ορίζεται η λογική λειτουργία ως NOT
        else:
            current_docs = set(index.get(term, [])) # Αναζήτηση του όρου στο ευρετήριο

            # Εφαρμογή της αντίστοιχης λογικής λειτουργίας
            if result_docs is None:  # Αν είναι ο πρώτος όρος στο ερώτημα...
                if operation == "NOT": # εάν η πρώτη λέξη καθιστά το operation NOT...
                    result_docs = all_docs - current_docs # Επιστρέφουμε τα έγγραφα που δεν περιέχουν τον όρο
                else:  # αλλιώς...  
                    result_docs = current_docs # Το result_docs αρχικοποιείται με τα έγγραφα του όρου
            elif operation == "AND": # Εάν η λειτουργία είναι AND... 
                result_docs.intersection_update(current_docs) # διατηρούνται μόνο τα κοινά έγγραφα μεταξύ των τρέχοντων και προηγούμενων αποτελεσμάτων
            elif operation == "OR": # Εάν η λειτουργία είναι OR...
                result_docs.update(current_docs) # προστίθενται τα νέα έγγραφα στα υπάρχοντα αποτελέσματα
            elif operation == "NOT": # Εάν η λειτουργία είναι NOT...
                result_docs.difference(current_docs) # αφαιρούνται τα νέα έγγραφα από τα αποτελέσματα

            operation = None  # Επαναφορά λειτουργίας για επόμενο όρο
    
    if result_docs: # Αν υπάρχουν αποτελέσματα...
        return sorted(list(result_docs)) # Επιστρέφονται ταξινομημένα ως λίστα
    else: # Αν δεν υπάρχουν αποτελέσματα...
        return [] # Επιστρέφεται κενή λίστα
```

Στη συνέχεια παρουσιάζεται ένα παράδειγμα χρήσης της συνάρτησης boolean_search, κατά το οποίο ζητάται από τον χρήστη να δώσει 4 ερωτήματα. Αν υπάρχουν έγγραφα(άρθρα) που να απαντούν στο ερώτημα του χρήστη, εκτυπώνονται στο τερματικό.

```python
# Παράδειγμα Χρήσης
i = 0  # Αρχικοποίηση μετρητή για τον αριθμό των ερωτημάτων
while i < 5:  # Επανάληψη για έως 4 ερωτήματα από τον χρήστη
    query = input("\nΔώσε το ερώτημα σου: ")  # Ζητά εισαγωγή ερωτήματος από τον χρήστη
    matching_docs = boolean_search(query, inverted_index)  # Αναζήτηση σχετικών εγγράφων μέσω της συνάρτησης boolean_search

    # Εμφάνιση των σχετικών εγγράφων
    if matching_docs:  # Αν βρέθηκαν σχετικά έγγραφα
        print(f"Τα έγγραφα που ταιριάζουν στο ερώτημα '{query}': {matching_docs}")  # Εμφάνιση των σχετικών εγγράφων
    else:  # Αν δεν βρέθηκαν σχετικά έγγραφα
        print(f"Δεν βρέθηκαν αποτελέσματα για το ερώτημα '{query}'")  # Μήνυμα ότι δεν υπάρχουν αποτελέσματα
    
    i += 1  # Αύξηση του μετρητή κατά 1 μετά από κάθε ερώτημα
```

### Ερώτημα β: Κατάταξη αποτελεσμάτων (Ranking)
Στο συγκεκριμένο ερώτημα θα εφαρμόσουμε έναν βασικό αλγόριθμο κατάταξης TF-IDF. Έπειτα θα υλοποιήσουμε ένα μενού με το οποίο ο χρήστης θα έχει τη δυνατότητα επιλογής ενός εκ των τριών αλγορίθμων ανάκτησης εγγράφων με βάση τα ερωτήματα του. Πο συγκεκριμένα θα υλοποιήσουμε τους αλγορίθμους Boolean Retrieval, Vector Space Model (VSM) και Probabilistic Retrieval Model(Okapi BM25).

#### Αλγόριθμος Κατάταξης TF-IDF
Σε αυτό το βήμα εφαρμόζεται ο αλγόριθμος TF-IDF (Term Frequency - Inverse Document Frequency) για την κατάταξη των αποτελεσμάτων αναζήτησης. Υλοποιήσαμε τον υπολογισμό του TF για κάθε όρο σε κάθε άρθρο και του IDF για τον προσδιορισμό της σημασίας των όρων στο σύνολο των εγγράφων. Στη συνέχεια, υπολογίστηκε το TF-IDF για κάθε ερώτημα του χρήστη και ταξινομήθηκαν τα αποτελέσματα με βάση το σκορ TF-IDF.

```python
import json
from sklearn.feature_extraction.text import TfidfVectorizer  # Εισάγουμε τον TfidfVectorizer από την βιβλιοθήκη sklearn για την υπολογισμό του TF-IDF
from nltk.tokenize import word_tokenize  # Εισάγουμε την συνάρτηση word_tokenize από την βιβλιοθήκη nltk για την επεξεργασία του κειμένου
import pandas as pd  # Εισάγουμε την βιβλιοθήκη pandas για την επεξεργασία και παρουσίαση των αποτελεσμάτων

# Φόρτωση των καθαρισμένων άρθρων από το αρχείο JSON
with open(r"C:\Users\07dig\OneDrive\structured_tokenized_articles.json", 'r') as f:
    articles = json.load(f)  # Διαβάζουμε τα δεδομένα του JSON και τα φορτώνουμε στη μεταβλητή articles

# Συνάρτηση για την κατάταξη των εγγράφων με βάση το TF-IDF
def my_tfidf(query, articles):
    # Προετοιμασία του corpus (συλλογή εγγράφων) και των ID των εγγράφων
    corpus = [" ".join(article['Body_cleaned']) for article in articles]  # Δημιουργούμε το corpus από τα tokens του Body_cleaned
    doc_ids = [article['ID'] for article in articles]  # Αποθηκεύουμε τα ID των εγγράφων

    # Βήμα 1: Εφαρμογή του TfidfVectorizer
    vectorizer = TfidfVectorizer()  # Δημιουργούμε έναν αντικείμενο TfidfVectorizer
    tfidf_matrix = vectorizer.fit_transform(corpus)  # Υπολογίζουμε τον πίνακα TF-IDF για όλα τα έγγραφα

    # Βήμα 2: Επεξεργασία και tokenization του ερωτήματος
    tokenized_query = " ".join(word_tokenize(query.lower()))  # Κάνουμε tokenization του ερωτήματος και το μετατρέπουμε σε πεζά
    query_vector = vectorizer.transform([tokenized_query])  # Μετατρέπουμε το ερώτημα σε μορφή TF-IDF χρησιμοποιώντας τον ίδιο vectorizer

    # Βήμα 3: Υπολογισμός των σκορ για όλα τα έγγραφα
    scores = (tfidf_matrix @ query_vector.T).toarray().flatten()  # Υπολογίζουμε τα σκορ (πιθανότητες) των εγγράφων βάσει του ερωτήματος

    # Βήμα 4: Ταξινόμηση και εμφάνιση των αποτελεσμάτων
    results = pd.DataFrame({
        'Document ID': doc_ids,  # Ταυτότητα του εγγράφου
        'TF-IDF Score': scores  # Σκορ TF-IDF για το έγγραφο
    }).sort_values(by='TF-IDF Score', ascending=False)  # Ταξινομούμε τα αποτελέσματα με βάση το σκορ TF-IDF από το υψηλότερο προς το χαμηλότερο

    # Εμφάνιση των 10 κορυφαίων αποτελεσμάτων
    if not results.empty:
        print("\nTop 10 Results:")  # Εκτυπώνουμε τον τίτλο "Top 10 αποτελέσματα"
        print(results.head(10))  # Εμφανίζουμε τα 10 κορυφαία αποτελέσματα
    else: 
        print("")  # Εάν δεν υπάρχουν αποτελέσματα, εκτυπώνουμε τίποτα
```

Παρακάτω παρουσιάζεται ένα παράδειγμα χρήσης του αλγορίθμου κατάταξης TF-IDF:

```python
# Παράδειγμα Χρήσης
query = input("\nΔώσε το ερώτημα σου: ")
my_tfidf(query, articles)
```

#### Αλγόριθμος Boolean Retrieval
Ο αλγόριθμος Boolean Retrieval είναι μια απλή και ισχυρή μέθοδος αναζήτησης, η οποία βασίζεται στην εφαρμογή λογικών πράξεων (AND, OR, NOT) για την ανάκτηση εγγράφων που περιέχουν συγκεκριμένους όρους ή συνδυασμούς αυτών. Η αναζήτηση πραγματοποιείται με βάση ένα αντεστραμμένο ευρετήριο, όπου οι όροι του κειμενικού περιεχομένου και πιο συγκεκριμένα τα tokens που έχουν εξαχθεί από το dataset στο βήμα 3 αντιστοιχούν σε λίστες εγγράφων που περιέχουν αυτούς τους όρους. Τέλος αν υπάρχουν έγγραφφα που ανακτώνται με βάση το ερώτημα του χρήστη, εκτυπώνονται στο τερματικό.

```python
import json
from nltk.tokenize import word_tokenize

# Φόρτωση του αντεστραμμένου ευρετηρίου ως λεξικό
with open(r"C:\Users\07dig\OneDrive\inverted_index.json", 'r') as f:
    inverted_index = json.load(f)

# Συνάρτηση για την εκτέλεση Boolean αναζήτησης
def boolean_retrieval(query, index):
    query_terms = word_tokenize(query.lower())  # Χωρισμός ερωτήματος σε λέξεις και μετατροπή σε πεζά
    result_docs = None  # Αρχικοποίηση με None
    all_docs = set()  # Σύνολο με όλα τα έγγραφα στο ευρετήριο
    operation = None # Αρχικά χωρίς λογική λειτουργία
    
     # Δημιουργία του συνόλου όλων των εγγράφων
    for docs in index.values():
        all_docs.update(docs)
    
    for term in query_terms:  # Επεξεργασία κάθε όρου στο ερώτημα
        if term in ["and", "AND"]: # Εάν ο όρος είναι and ή AND...
            operation = "AND" # Ορίζεται η λογική λειτουργία ως AND
        elif term in ["or", "OR"]: # Εάν ο όρος είναι or ή OR...
            operation = "OR" # Ορίζεται η λογική λειτουργία ως OR
        elif term in ["not", "NOT"]: # Εάν ο όρος είναι not ή NOT...
            operation = "NOT" # Ορίζεται η λογική λειτουργία ως NOT
        else:
            current_docs = set(index.get(term, [])) # Αναζήτηση του όρου στο ευρετήριο

            # Εφαρμογή της αντίστοιχης λογικής λειτουργίας
            if result_docs is None:  # Αν είναι ο πρώτος όρος στο ερώτημα...
                if operation == "NOT": # εάν η πρώτη λέξη καθιστά το operation NOT...
                    result_docs = all_docs - current_docs # Επιστρέφουμε τα έγγραφα που δεν περιέχουν τον όρο
                else:  # αλλιώς...  
                    result_docs = current_docs # Το result_docs αρχικοποιείται με τα έγγραφα του όρου
            elif operation == "AND": # Εάν η λειτουργία είναι AND... 
                result_docs.intersection_update(current_docs) # διατηρούνται μόνο τα κοινά έγγραφα μεταξύ των τρέχοντων και προηγούμενων αποτελεσμάτων
            elif operation == "OR": # Εάν η λειτουργία είναι OR...
                result_docs.update(current_docs) # προστίθενται τα νέα έγγραφα στα υπάρχοντα αποτελέσματα
            elif operation == "NOT": # Εάν η λειτουργία είναι NOT...
                result_docs.difference(current_docs) # αφαιρούνται τα νέα έγγραφα από τα αποτελέσματα

            operation = None  # Επαναφορά λειτουργίας για επόμενο όρο
    
    if result_docs: # Αν υπάρχουν αποτελέσματα...
        results = sorted(list(result_docs)) # Επιστρέφονται ταξινομημένα ως λίστα
    else: # Αν δεν υπάρχουν αποτελέσματα...
        results = [] # Επιστρέφεται κενή λίστα

    if results:
        print(f"Τα έγγραφα που ταιριάζουν στο ερώτημα '{query}': {results}")
    else:
        print(f"Δεν βρέθηκαν αποτελέσματα για το ερώτημα '{query}'")
```

Παρακάτω παρουσιάζεται ένα παράδειγμα χρήσης του αλγορίθμου Boolean Retrieval:

```python
# Παράδειγμα Χρήσης
query = input("\nΔώσε το ερώτημα σου: ")
boolean_retrieval(query, inverted_index)
```

#### Αλγόριθμος Vector Space Model (VSM)
Το Vector Space Model(VSM) βασίζεται σε διανυσματική αναπαράσταση των εγγράφων και του ερωτήματος. Υπολογίζει την ομοιότητα συνημιτόνου μεταξύ του διανύσματος του ερωτήματος και των διανυσμάτων των εγγράφων. Τέλος ταξινομούνται τα έγγραφα με βάση τη συνάφεια(ομοιότητα).

```python
# Εισαγωγή Βιοβλιοθηκων
import json 
import nltk 
from sklearn.feature_extraction.text import TfidfVectorizer  
from sklearn.metrics.pairwise import cosine_similarity  
import pandas as pd 

# Κατεβάζουμε τον tokenizer που είναι απαραίτητος για την επεξεργασία κειμένου από το nltk.
nltk.download('punkt')
from nltk.tokenize import word_tokenize  # Εισαγωγή της μεθόδου word_tokenize για διαχωρισμό λέξεων σε προτάσεις.

# Φορτώνουμε τα προεπεξεργασμένα άρθρα από ένα αρχείο JSON.
with open(r"C:\Users\07dig\OneDrive\structured_tokenized_articles.json", 'r') as f:
    articles = json.load(f)  # Διαβάζουμε τα δεδομένα από το αρχείο JSON και τα αποθηκεύουμε στη μεταβλητή `articles`.

# Υλοποίηση του Vector Space Model (VSM).
def vector_space_model(query, articles):
    # Προετοιμάζουμε το corpus (συλλογή κειμένων) και τις ταυτότητες των εγγράφων.
    documents = [' '.join(article['Body_cleaned']) for article in articles]  # Ενώνουμε το καθαρισμένο κειμενικό περιεχόμενο των άρθρων.
    doc_ids = [article['ID'] for article in articles]  # Αποθηκεύουμε τα ID των άρθρων.

    # Χωρίζουμε την ερώτηση σε λέξεις (tokens).
    tokenized_query = query.lower().split  # Μετατρέπουμε το ερώτημα σε πεζά γράμματα και το χωρίζουμε σε tokens.
    
    # Βήμα 1: Υπολογίζουμε την TF-IDF.
    vectorizer = TfidfVectorizer()  # Δημιουργούμε ένα αντικείμενο TfidfVectorizer.
    tfidf_matrix = vectorizer.fit_transform(documents)  # Υπολογίζουμε τον πίνακα TF-IDF για τα έγγραφα.
    query_vector = vectorizer.transform([query])  # Μετατρέπουμε το ερώτημα σε διανυσματική αναπαράσταση.

    # Βήμα 2: Υπολογίζουμε την ομοιότητα συνημιτόνου.
    cosine_similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()  # Υπολογίζουμε τις ομοιότητες με το συνημίτονο.

    # Βήμα 3: Δημιουργούμε DataFrame αποτελεσμάτων και τα ταξινομούμε.
    results = pd.DataFrame({
        'Document ID': doc_ids,  # Προσθέτουμε τα ID των εγγράφων.
        'Similarity Score': cosine_similarities  # Προσθέτουμε τα σκορ ομοιότητας.
    }).sort_values(by='Similarity Score', ascending=False)  # Ταξινομούμε κατά φθίνουσα σειρά σκορ.

    # Βήμα 4: Εμφανίζουμε τα κορυφαία 10 αποτελέσματα.
    if not results.empty:  # Ελέγχουμε αν υπάρχουν αποτελέσματα.
        print("\nTop 10 Results:")  # Εμφανίζουμε μήνυμα.
        print(results.head(10))  # Εμφανίζουμε τα 10 πρώτα αποτελέσματα.
    else:
        print("Δεν βρέθηκαν αποτελέσματα.")  # Μήνυμα αν δεν υπάρχουν αποτελέσματα.
```

Παρακάτω παρουσιάζεται ένα παράδειγμα χρήσης του αλγορίθμου Vector Space Model:

```python
# Παράδειγμα χρήσης 
query = input("Δώσε το ερώτημα σου: ")
vector_space_model(query, articles)
```

#### Αλγόριθμος  Probabilistic Retrieval Model (Okapi BM25)
Ο OkapiBM25 είναι μια μέθοδος ανάκτησης πληροφορίας που χρησιμοποιούμε για την αξιολόγηση της σχετικότητας των εγγράφων του συνόλου των δεδομένων με βάση το ερώτημα του χρήστη

Έπειτα προχωράμε στην υλοποίηση του αλγορίθμου ανάκτησης OkapiBM25

```python
# Εισαγωγή βιβλιοθηκων
import json 
from rank_bm25 import BM25Okapi  
from nltk.tokenize import word_tokenize  
import pandas as pd  

# Φόρτωση των προεπεξεργασμένων άρθρων από το αρχείο JSON.
with open(r"C:\Users\07dig\OneDrive\structured_tokenized_articles.json", 'r') as f:
    articles = json.load(f)  # Διαβάζουμε τα δεδομένα από το αρχείο JSON και τα αποθηκεύουμε στη μεταβλητή `articles`.

# Ορισμός συνάρτησης για την υλοποίηση του αλγορίθμου Okapi BM25.
def okapi_bm25(query, articles):
    # Δημιουργία λίστας με τους όρους κάθε άρθρου από το πεδίο `Body_cleaned`.
    documents = [article['Body_cleaned'] for article in articles]  # Τα σώματα των άρθρων σε μορφή λίστας όρων.
    # Δημιουργία λίστας με τα ID των άρθρων.
    doc_ids = [article['ID'] for article in articles]  # Τα ID των άρθρων.

    # Διαχωρισμός του ερωτήματος σε λέξεις (tokens).
    tokenized_query = query.split(" ")  # Χωρίζουμε το ερώτημα σε ξεχωριστές λέξεις.

    # Δημιουργία αντικειμένου BM25 με τα tokenized έγγραφα.
    bm25 = BM25Okapi(documents)  # Αρχικοποιούμε τον αλγόριθμο BM25 με τα έγγραφα.

    # Υπολογισμός των βαθμολογιών BM25 για το ερώτημα.
    doc_scores = bm25.get_scores(tokenized_query)  # Υπολογίζουμε τη βαθμολογία BM25 για κάθε έγγραφο.

    # Λήψη των κορυφαίων 10 εγγράφων που ταιριάζουν στο ερώτημα.
    top_n_docs = bm25.get_top_n(tokenized_query, documents, n=10)  # Εντοπίζουμε τα 10 κορυφαία έγγραφα.

    # Εύρεση των θέσεων των εγγράφων στο αρχικό corpus.
    top_indices = [documents.index(doc) for doc in top_n_docs]  # Εντοπίζουμε τις θέσεις των εγγράφων.

    # Δημιουργία DataFrame για την παρουσίαση των αποτελεσμάτων.
    results = pd.DataFrame({
        'Document ID': [doc_ids[i] for i in top_indices],  # Αντιστοίχιση των ID στα κορυφαία έγγραφα.
        'BM25 Score': [doc_scores[i] for i in top_indices]  # Αντιστοίχιση των βαθμολογιών BM25.
    }).sort_values(by='BM25 Score', ascending=False)  # Ταξινόμηση κατά φθίνουσα σειρά βαθμολογιών.

    # Εμφάνιση των αποτελεσμάτων.
    if not results.empty:  # Έλεγχος αν υπάρχουν αποτελέσματα.
        print("\nTop 10 Results:")  # Εμφάνιση μηνύματος για τα κορυφαία αποτελέσματα.
        print(results.head(10))  # Εμφάνιση των πρώτων 10 εγγράφων.
    else:
        print("Δεν βρέθηκαν αποτελέσματα.")  # Μήνυμα όταν δεν υπάρχουν αποτελέσματα.
```

```python
import os  # Διαχείριση αρχείων και διαδρομών
import nltk  # Επεξεργασία φυσικής γλώσσας
from sklearn.feature_extraction.text import TfidfVectorizer  # Υπολογισμός TF-IDF
from sklearn.metrics.pairwise import cosine_similarity  # Υπολογισμός ομοιότητας συνημιτόνου
from rank_bm25 import BM25Okapi  # Αλγόριθμος BM25
import pandas as pd  # Διαχείριση δεδομένων
import kagglehub  # Λήψη δεδομένων από το Kaggle

# Λήψη των απαραίτητων πόρων του NLTK
nltk.download('punkt')
from nltk.tokenize import word_tokenize  # Διαχωρισμός λέξεων σε tokens

# Κατέβασμα του συνόλου δεδομένων μέσω του kagglehub
path = kagglehub.dataset_download("dmaso01dsta/cisi-a-dataset-for-information-retrieval")

# 1. Ανάλυση συνόλου δεδομένων
def parse_cisi_dataset(cisi_all, cisi_qry, cisi_rel):
    """Ανάλυση αρχείων του συνόλου δεδομένων CISI."""
    # Ανάλυση αρχείου CISI.ALL (έγγραφα)
    documents = {}
    with open(cisi_all, 'r') as f:
        lines = ""
        for l in f.readlines():
            lines += "\n" + l.strip() if l.startswith(".") else " " + l.strip()
        lines = lines.lstrip("\n").split("\n")
        doc_id = None
        doc_text = ""
        for l in lines:
            if l.startswith(".I"):  # Νέο έγγραφο
                if doc_id is not None:  # Αποθήκευση προηγούμενου εγγράφου
                    documents[doc_id] = doc_text.lstrip(" ")
                doc_id = l.split(" ")[1].strip()
                doc_text = ""
            elif l.startswith(".X"):  # Τέλος εγγράφου
                documents[doc_id] = doc_text.lstrip(" ")
                doc_id = None
                doc_text = ""
            else:  # Συνέχιση κειμένου εγγράφου
                doc_text += l.strip()[3:] + " "

    # Ανάλυση αρχείου CISI.QRY (ερωτήματα)
    queries = {}
    with open(cisi_qry, 'r') as f:
        lines = ""
        for l in f.readlines():
            lines += "\n" + l.strip() if l.startswith(".") else " " + l.strip()
        lines = lines.lstrip("\n").split("\n")
        qry_id = None
        for l in lines:
            if l.startswith(".I"):  # Νέο ερώτημα
                qry_id = l.split(" ")[1].strip()
            elif l.startswith(".W"):  # Κείμενο ερωτήματος
                queries[qry_id] = l.strip()[3:]
                qry_id = None

    # Ανάλυση αρχείου CISI.REL (σχετικότητα)
    relevance = {}
    with open(cisi_rel, 'r') as f:
        for l in f.readlines():
            parts = l.split()
            qry_id = parts[0]
            doc_id = parts[1]
            if qry_id in relevance:  # Προσθήκη εγγράφου στη σχετικότητα
                relevance[qry_id].append(doc_id)
            else:  # Δημιουργία νέας καταχώρισης για το ερώτημα
                relevance[qry_id] = [doc_id]

    return documents, queries, relevance

# Φόρτωση αρχείων του συνόλου δεδομένων CISI
cisi_all_path = os.path.join(path, "CISI.ALL")
cisi_qry_path = os.path.join(path, "CISI.QRY")
cisi_rel_path = os.path.join(path, "CISI.REL")

documents, queries, relevance = parse_cisi_dataset(cisi_all_path, cisi_qry_path, cisi_rel_path)

# Επιβεβαίωση της δομής των δεδομένων
print(f"Number of Documents: {len(documents)}")  # Αριθμός εγγράφων
print(f"Number of Queries: {len(queries)}")  # Αριθμός ερωτημάτων
print(f"Number of Query-Relevance Pairs: {len(relevance)}")  # Αριθμός ζευγών σχετικότητας
```

#### Αλγόριθμος TF-IDF και η αξιολόγηση του
Παρακάτω υλοποιείται ο αλγόριθμος κατάταξης TF-IDF με την ίδια λογική που φτιάχτηκε στο βήμα 2. Αρχικά χρησιμοποιεί τα tokenized έγγραφα και ερωτήματα που φορτώνονται από αρχεία JSON. Αρχικά, δημιουργούνται τα corpus για τα έγγραφα και τα ερωτήματα. Στη συνέχεια, χρησιμοποιείται ο TfidfVectorizer για τον υπολογισμό του πίνακα TF-IDF των εγγράφων και τη μετατροπή των ερωτημάτων σε διανύσματα TF-IDF. Οι βαθμολογίες TF-IDF υπολογίζονται για κάθε έγγραφο σε σχέση με κάθε ερώτημα και τα αποτελέσματα ταξινομούνται κατά φθίνουσα σειρά. Τέλος, επιστρέφονται τα 10 κορυφαία αποτελέσματα για κάθε ερώτημα.

```python
######################### TF-IDF Κατάταξη Εγγράφων ########################
import json
from sklearn.feature_extraction.text import TfidfVectorizer  # Εισαγωγή του TfidfVectorizer από το scikit-learn για υπολογισμό TF-IDF
from nltk.tokenize import word_tokenize  # Εισαγωγή του word_tokenize από το NLTK για διαχωρισμό λέξεων
import pandas as pd  # Εισαγωγή της βιβλιοθήκης pandas για οργάνωση και εμφάνιση αποτελεσμάτων

# Φόρτωση των tokenized εγγράφων
with open(r"C:\Users\07dig\OneDrive\tokenized_documents.json", 'r') as f:
    tokenized_docs = json.load(f)  # Φόρτωση των tokenized εγγράφων από το JSON

# Δημιουργία corpus και doc_ids από τα tokenized έγγραφα
corpus = [" ".join(doc) for doc in tokenized_docs.values()]  # Ένωση των tokens κάθε εγγράφου σε μια ενιαία συμβολοσειρά
doc_ids = list(tokenized_docs.keys())  # Εξαγωγή των ID των εγγράφων

# Φόρτωση των tokenized ερωτημάτων
with open(r"C:\Users\07dig\OneDrive\tokenized_queries.json", 'r') as f:
    tokenized_queries = json.load(f)  # Φόρτωση των tokenized ερωτημάτων από το JSON

# Δημιουργία corpus και query_ids από τα tokenized ερωτήματα
query_corpus = [" ".join(query) for query in tokenized_queries.values()]  # Ένωση των tokens κάθε ερωτήματος
query_ids = list(tokenized_queries.keys())  # Εξαγωγή των ID των ερωτημάτων

# Ορισμός της συνάρτησης my_tfidf_search
def my_tfidf_search(tokenized_queries, tokenized_docs):
    # Βήμα 1: Εφαρμογή του TfidfVectorizer
    vectorizer = TfidfVectorizer()  # Δημιουργία αντικειμένου TfidfVectorizer
    tfidf_matrix = vectorizer.fit_transform(corpus)  # Υπολογισμός του πίνακα TF-IDF για όλα τα έγγραφα

    # Βήμα 2: Επεξεργασία και μετατροπή των ερωτημάτων
    query_vectors = vectorizer.transform(query_corpus)  # Μετατροπή όλων των ερωτημάτων σε διανύσματα TF-IDF με τον ίδιο vectorizer

    # Βήμα 3: Υπολογισμός βαθμολογιών για όλα τα έγγραφα για κάθε ερώτημα
    results = []  # Αρχικοποίηση μιας κενής λίστας για αποθήκευση των αποτελεσμάτων

    for query_id, query_vector in zip(query_ids, query_vectors):
        scores = (tfidf_matrix @ query_vector.T).toarray().flatten()  # Υπολογισμός βαθμολογιών για κάθε έγγραφο σε σχέση με το ερώτημα
        tfidf_results = pd.DataFrame({
            'Document ID': doc_ids,  # ID εγγράφου
            'TF-IDF Score': scores  # Αντίστοιχες βαθμολογίες TF-IDF για κάθε έγγραφο
        }).sort_values(by='TF-IDF Score', ascending=False)  # Ταξινόμηση κατά φθίνουσα σειρά βαθμολογιών TF-IDF

        # Προσθήκη των κορυφαίων 10 αποτελεσμάτων για το ερώτημα στη λίστα αποτελεσμάτων
        results.append({
            'Query ID': query_id,
            'Top 10 Results': tfidf_results.head(10),
        })

    return results  # Επιστροφή των αποτελεσμάτων για όλα τα ερωτήματα

# Απόκτηση των αποτελεσμάτων TF-IDF για όλα τα ερωτήματα
tfidf_results = my_tfidf_search(tokenized_queries, tokenized_docs)

# Εμφάνιση των αποτελεσμάτων για κάθε ερώτημα
for tfidf_result in tfidf_results:
    print(f"\nQuery ID: {tfidf_result['Query ID']}")
    print("Top 10 Results:")
    print(tfidf_result['Top 10 Results'])  # Εμφάνιση των κορυφαίων 10 αποτελεσμάτων για κάθε ερώτημα
```

Έπειτα θα προχωρήσουμε στην αξιολόγηση της απόδοσης για τον αλγόριθμο κατάταξης TF-IDF χρησιμοποιώντας τις εξής μετρικές: 
- Ακρίβεια (Precision)
- Ανάκληση (Recall)
- F1-score 
- Μέση Ακρίβεια Ερωτήματος (AP)
- Μέση Ακρίβεια Ερωτημάτων (MAP)

Μεθοδοι Αξιολόγησης:
- Υλοποιούμε τη συνάρτηση evaluate_ranking,  με την οποία πραγματοποιούμε evaluation για κάθε ερώτημα για τον αλγόριθμο κατάταξης TF-IDF και υπολογίζονται η μετρικές precision, recall, F1-score και AP.
- Υλοποιούμε την συνάρτηση evaluate_all_queries, με την οποία πραγματοποιούμε evaluation για όλα τα ερωτήματα για τον αλγόριθμο κατάταξης TF-IDF υπολογίζονται οι μέσες τιμές για precision, recall, F1-score καθώς επίσης υπολογίζεται η μέση ακρίβεια (MAP). Η MAP ορίζεται ως η μέση τιμή της AP κάθε ερωτήματος.

```python
from sklearn.metrics import precision_score, recall_score, f1_score  # Εισαγωγή μετρικών από τη βιβλιοθήκη scikit-learn
import numpy as np  # Εισαγωγή της βιβλιοθήκης numpy για υπολογισμούς

# Συνάρτηση για την αξιολόγηση της κατάταξης για ένα ερώτημα
def evaluate_ranking(query_id, ranked_docs, relevance):
    """Αξιολόγηση ακρίβειας (precision), ανάκλησης (recall), F1-score και μέσης ακρίβειας (AP) για ένα ερώτημα."""
    query_id = int(query_id)  # Μετατροπή του query_id σε ακέραιο
    relevant_docs = set(map(int, relevance.get(query_id, [])))  # Εύρεση σχετικών εγγράφων και μετατροπή τους σε ακέραιους
    retrieved_docs = list(map(int, ranked_docs))  # Μετατροπή των ανακτημένων εγγράφων σε ακέραιους
    retrieved_relevant = [doc for doc in retrieved_docs if doc in relevant_docs]  # Εύρεση των σχετικών εγγράφων που ανακτήθηκαν

    # Υπολογισμός ακρίβειας (precision), ανάκλησης (recall), και F1-score
    precision = len(retrieved_relevant) / len(retrieved_docs) if retrieved_docs else 0
    recall = len(retrieved_relevant) / len(relevant_docs) if relevant_docs else 0
    f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    # Υπολογισμός Μέσης Ακρίβειας (Average Precision - AP)
    ap = 0
    if relevant_docs:
        relevant_hits = 0
        for i, doc in enumerate(retrieved_docs, start=1):  # Για κάθε έγγραφο στη λίστα
            if doc in relevant_docs:
                relevant_hits += 1
                ap += relevant_hits / i  # Ακρίβεια στη συγκεκριμένη θέση
        ap /= len(relevant_docs)

    return precision, recall, f1, ap  # Επιστροφή των υπολογισμένων μετρικών

# Συνάρτηση για την αξιολόγηση όλων των ερωτημάτων
def evaluate_all_queries(tfidf_results, relevance):
    """Αξιολόγηση όλων των ερωτημάτων μέσω της `evaluate_ranking`."""
    overall_precision, overall_recall, overall_f1, average_precisions = [], [], [], []  # Λίστες για αποθήκευση των μετρικών

    for result in tfidf_results:
        query_id = int(result['Query ID'])  # Μετατροπή του query_id σε ακέραιο
        ranked_docs = list(map(int, result['Top 10 Results']['Document ID'].tolist()))  # Εξαγωγή των ανακτημένων εγγράφων

        # Αξιολόγηση της κατάταξης για το συγκεκριμένο ερώτημα
        precision, recall, f1, ap = evaluate_ranking(query_id, ranked_docs, relevance)

        # Εμφάνιση των αποτελεσμάτων για το ερώτημα
        print(f"\nQuery ID: {query_id}")
        print(f"Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}, Average Precision: {ap:.4f}")

        # Προσθήκη των μετρικών στις συνολικές λίστες
        overall_precision.append(precision)
        overall_recall.append(recall)
        overall_f1.append(f1)
        average_precisions.append(ap)

    # Υπολογισμός μέσων τιμών για όλες τις μετρικές
    mean_precision = np.mean(overall_precision)
    mean_recall = np.mean(overall_recall)
    mean_f1 = np.mean(overall_f1)
    map_score = np.mean(average_precisions)

    # Εμφάνιση συνολικών αποτελεσμάτων
    print("\n=== Overall Evaluation ===")
    print(f"Mean Precision: {mean_precision:.4f}")
    print(f"Mean Recall: {mean_recall:.4f}")
    print(f"Mean F1-Score: {mean_f1:.4f}")
    print(f"Mean Average Precision (MAP): {map_score:.4f}")

# Βεβαιωθείτε ότι όλα τα IDs στο relevance είναι ακέραιοι κατά τη φόρτωση
relevance = {int(qid): list(map(int, docs)) for qid, docs in relevance.items()}

# Αξιολόγηση των αποτελεσμάτων TF-IDF για όλα τα ερωτήματα
evaluate_all_queries(tfidf_results, relevance)
```

#### Αλγόριθμος Boolean Retrieval και η αξιολόγηση του
Παρακάτω υλοποιείται o αλγόριθμος ανάκτησης Boolean Retrieval για τα ερωτήματα με την ίδια λογική που φτιάχτηκε στο βήμα 2. Αρχικά, φορτώνονται τα επεξεργασμένα ερωτήματα από το αρχείο JSON. Στη συνέχεια, για κάθε ερώτημα, η συνάρτηση boolean_retrieval εκτελεί αναζήτηση με τους όρους του ερωτήματος, εφαρμόζοντας λογικές λειτουργίες (AND, OR, NOT) στους όρους. Τα αποτελέσματα ταξινομούνται και επιστρέφονται τα 10 πρώτα για κάθε ερώτημα. Τέλος, εμφανίζονται τα αποτελέσματα για κάθε ερώτημα, περιλαμβάνοντας τα κορυφαία 10 αποτελέσματα και όλα τα αποτελέσματα που ανακτήθηκαν.

```python
import json  # Εισαγωγή της βιβλιοθήκης για χειρισμό JSON αρχείων

# Φόρτωση των επεξεργασμένων ερωτημάτων
with open(r"C:\Users\07dig\OneDrive\tokenized_queries.json", 'r') as f:
    tokenized_queries = json.load(f)  # Φορτώνουμε τα tokenized ερωτήματα από το JSON αρχείο

# Συνάρτηση για Boolean retrieval
def boolean_retrieval(query_terms, index):
    """Εκτέλεση Boolean αναζήτησης για ένα μόνο ερώτημα."""
    result_docs = None  # Αρχικοποίηση συνόλου αποτελεσμάτων
    all_docs = set()  # Όλα τα IDs εγγράφων στο ευρετήριο
    operation = None  # Λογική λειτουργία

    # Δημιουργία συνόλου όλων των εγγράφων στο ευρετήριο
    for docs in index.values():
        all_docs.update(docs)

    for term in query_terms:  # Επεξεργασία κάθε όρου του ερωτήματος
        if term.lower() in ["and"]:  # AND λειτουργία
            operation = "AND"
        elif term.lower() in ["or"]:  # OR λειτουργία
            operation = "OR"
        elif term.lower() in ["not"]:  # NOT λειτουργία
            operation = "NOT"
        else:
            current_docs = set(index.get(term, []))  # Εύρεση εγγράφων για τον συγκεκριμένο όρο

            # Εφαρμογή λογικών λειτουργιών
            if result_docs is None:  # Πρώτος όρος στο ερώτημα
                result_docs = all_docs - current_docs if operation == "NOT" else current_docs
            elif operation == "AND":  # Λογική AND
                result_docs.intersection_update(current_docs)
            elif operation == "OR":  # Λογική OR
                result_docs.update(current_docs)
            elif operation == "NOT":  # Λογική NOT
                result_docs.difference_update(current_docs)

            operation = None  # Επαναφορά λειτουργίας για τον επόμενο όρο

    # Επιστροφή ταξινομημένων αποτελεσμάτων
    return sorted(result_docs) if result_docs else []

# Επεξεργασία κάθε ερωτήματος μέσω Boolean αναζήτησης
boolean_results = []
for query_id, query_terms in tokenized_queries.items():
    matching_docs = boolean_retrieval(query_terms, inverted_index)  # Εκτέλεση Boolean retrieval
    boolean_results.append({
        "Query ID": query_id,
        "Top 10 Results": matching_docs[:10],  # Τα κορυφαία 10 αποτελέσματα
        "Ranked Docs": matching_docs  # Όλα τα αποτελέσματα
    })

# Εμφάνιση αποτελεσμάτων για κάθε ερώτημα
for boolean_result in boolean_results:
    print(f"\nQuery ID: {boolean_result['Query ID']}")
    print(boolean_result['Top 10 Results'])  # Εμφάνιση των κορυφαίων 10 αποτελεσμάτων
    print("Ranked Docs:")
    print(boolean_result['Ranked Docs'])  # Εμφάνιση όλων των εγγράφων που ανακτήθηκαν
```

- Ακρίβεια (Precision)
- Ανάκληση (Recall)
- F1-score 
- Μέση Ακρίβεια Ερωτήματος (AP)
- Μέση Ακρίβεια Ερωτημάτων (MAP)

Μεθοδοι Αξιολόγησης:
- Υλοποιούμε τη συνάρτηση evaluate_boolean_ranking,  με την οποία πραγματοποιούμε evaluation για κάθε ερώτημα για τον αλγόριθμο ανάκτησης Boolean Retrieval και υπολογίζονται η μετρικές precision, recall, F1-score και AP.
- Υλοποιούμε την συνάρτηση evaluate_all_boolean_queries, με την οποία πραγματοποιούμε evaluation για όλα τα ερωτήματα για τον αλγόριθμο ανάκτησης Boolean Retrieval υπολογίζονται οι μέσες τιμές για precision, recall, F1-score καθώς επίσης υπολογίζεται η μέση ακρίβεια (MAP). Η MAP ορίζεται ως η μέση τιμή της AP κάθε ερωτήματος.

```python
import numpy as np  # Εισαγωγή της βιβλιοθήκης numpy για αριθμητικούς υπολογισμούς

def evaluate_boolean_ranking(query_id, retrieved_docs, relevance):
    """Αξιολόγηση precision, recall, F1-score και AP για Boolean ερώτημα με όλα τα ανακτημένα έγγραφα."""
    query_id = int(query_id)  # Εξασφάλιση ότι το query ID είναι ακέραιος αριθμός
    relevant_docs = set(map(int, relevance.get(query_id, [])))  # Μετατροπή των σχετικών εγγράφων σε ακέραιους
    retrieved_docs = list(map(int, retrieved_docs))  # Μετατροπή των ανακτημένων εγγράφων σε ακέραιους
    retrieved_relevant = [doc for doc in retrieved_docs if doc in relevant_docs]  # Ανακτώμενα σχετικά έγγραφα

    # Υπολογισμός precision, recall, F1-score
    precision = len(retrieved_relevant) / len(retrieved_docs) if retrieved_docs else 0
    recall = len(retrieved_relevant) / len(relevant_docs) if relevant_docs else 0
    f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    # Υπολογισμός Average Precision (AP)
    ap = 0
    if relevant_docs:
        relevant_hits = 0
        for i, doc in enumerate(retrieved_docs, start=1):
            if doc in relevant_docs:
                relevant_hits += 1
                ap += relevant_hits / i  # Precision σε κάθε relevant hit
        ap /= len(relevant_docs)

    return precision, recall, f1, ap


def evaluate_all_boolean_queries(boolean_results, relevance):
    """Αξιολόγηση όλων των Boolean ερωτημάτων και υπολογισμός MAP με όλα τα ανακτημένα έγγραφα."""
    overall_precision, overall_recall, overall_f1, average_precisions = [], [], [], []  # Αρχικοποίηση λιστών μετρικών

    for result in boolean_results:
        query_id = int(result['Query ID'])  # Μετατροπή του Query ID σε ακέραιο
        retrieved_docs = list(map(int, result['Ranked Docs']))  # Ανακτημένα έγγραφα για το συγκεκριμένο ερώτημα

        # Αξιολόγηση κατάταξης για το συγκεκριμένο ερώτημα
        precision, recall, f1, ap = evaluate_boolean_ranking(query_id, retrieved_docs, relevance)

        # Εκτύπωση αποτελεσμάτων για το ερώτημα
        print(f"\nQuery ID: {query_id}")
        print(f"Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}, Average Precision: {ap:.4f}")

        # Προσθήκη μετρικών στις συνολικές λίστες
        overall_precision.append(precision)
        overall_recall.append(recall)
        overall_f1.append(f1)
        average_precisions.append(ap)

    # Υπολογισμός συνολικών μετρικών
    mean_precision = np.mean(overall_precision)
    mean_recall = np.mean(overall_recall)
    mean_f1 = np.mean(overall_f1)
    map_score = np.mean(average_precisions)

    # Εκτύπωση συνολικών αποτελεσμάτων
    print("\n=== Overall Evaluation ===")
    print(f"Mean Precision: {mean_precision:.4f}")
    print(f"Mean Recall: {mean_recall:.4f}")
    print(f"Mean F1-Score: {mean_f1:.4f}")
    print(f"Mean Average Precision (MAP): {map_score:.4f}")

# Μετατροπή όλων των IDs σε ακέραιους στο relevance
relevance = {int(qid): list(map(int, docs)) for qid, docs in relevance.items()}

# Αξιολόγηση των αποτελεσμάτων Boolean
evaluate_all_boolean_queries(boolean_results, relevance)
```

#### Αλγόριθμος Vector Space Model (VSM) και η αξιολόγηση του
Παρακάτω υλοποιείται η διαδικασία κατάταξης εγγράφων με τη μέθοδο Vector Space Model (VSM) για τα ερωτήματα με την ίδια λογική που φτιάχτηκε στο βήμα 2. Αρχικά, φορτώνονται τα επεξεργασμένα έγγραφα και ερωτήματα από τα αρχεία JSON. Στη συνέχεια, για κάθε ερώτημα, η συνάρτηση vector_space_model υπολογίζει την ομοιότητα μεταξύ των εγγράφων και του ερωτήματος χρησιμοποιώντας την ομοιότητα συνημιτόνου (cosine similarity). Ο αλγόριθμος ταξινομεί τα αποτελέσματα κατά φθίνουσα σειρά ομοιότητας και επιστρέφει τα κορυφαία 10 αποτελέσματα για κάθε ερώτημα. Τέλος, εμφανίζονται τα αποτελέσματα για κάθε ερώτημα, περιλαμβάνοντας τα κορυφαία 10 αποτελέσματα και όλες τις βαθμολογίες ομοιότητας.

```python
import json  # Εισαγωγή της βιβλιοθήκης JSON για διαχείριση αρχείων JSON
import nltk  # Εισαγωγή της βιβλιοθήκης nltk 
from sklearn.feature_extraction.text import TfidfVectorizer  # Υπολογισμός TF-IDF
from sklearn.metrics.pairwise import cosine_similarity  # Υπολογισμός ομοιότητας συνημιτόνου
import pandas as pd  # Εισαγωγή της pandas για επεξεργασία και παρουσίαση δεδομένων

# Λήψη του απαραίτητου tokenizer του NLTK
nltk.download('punkt')

# Φόρτωση των προεπεξεργασμένων εγγράφων και ερωτημάτων
with open(r"C:\Users\07dig\OneDrive\tokenized_documents.json", 'r') as f:
    tokenized_docs = json.load(f)  # Φορτώνουμε τα εγγράφων από το αρχείο JSON

with open(r"C:\Users\07dig\OneDrive\tokenized_queries.json", 'r') as f:
    tokenized_queries = json.load(f)  # Φορτώνουμε τα ερωτήματα από το αρχείο JSON

# Προετοιμασία του corpus των εγγράφων
doc_corpus = [' '.join(tokens) for tokens in tokenized_docs.values()]  # Δημιουργία corpus ενώνοντας τα tokens κάθε εγγράφου
doc_ids = list(tokenized_docs.keys())  # Λίστα με τα IDs των εγγράφων

# Υλοποίηση Vector Space Model (VSM)
def vector_space_model(query, doc_corpus, doc_ids):
    """Κατάταξη εγγράφων με χρήση του Vector Space Model."""
    # Βήμα 1: Υπολογισμός TF-IDF
    vectorizer = TfidfVectorizer()  # Δημιουργία αντικειμένου TfidfVectorizer
    tfidf_matrix = vectorizer.fit_transform(doc_corpus)  # Υπολογισμός διανυσμάτων εγγράφων
    query_vector = vectorizer.transform([query])  # Δημιουργία διανύσματος για το ερώτημα

    # Βήμα 2: Υπολογισμός ομοιότητας συνημιτόνου
    cosine_similarities = cosine_similarity(query_vector, tfidf_matrix).flatten()  # Ομοιότητες με το ερώτημα

    # Βήμα 3: Δημιουργία και ταξινόμηση των αποτελεσμάτων
    results = pd.DataFrame({
        'Document ID': doc_ids,  # Ταυτότητα εγγράφου
        'Similarity Score': cosine_similarities  # Βαθμολογία ομοιότητας
    }).sort_values(by='Similarity Score', ascending=False)  # Ταξινόμηση κατά φθίνουσα σειρά

    # Βήμα 4: Επιστροφή των κορυφαίων 10 αποτελεσμάτων
    return results.head(10)

# Επεξεργασία κάθε ερωτήματος με χρήση του Vector Space Model
vsm_results = []
for query_id, query_tokens in tokenized_queries.items():
    query = ' '.join(query_tokens)  # Μετατροπή του tokenized ερωτήματος σε string
    top_results = vector_space_model(query, doc_corpus, doc_ids)  # Εφαρμογή του αλγορίθμου VSM

    # Προσθήκη αποτελεσμάτων για το συγκεκριμένο ερώτημα
    vsm_results.append({
        "Query ID": query_id,  # Ταυτότητα ερωτήματος
        "Top 10 Results": top_results  # Κορυφαία 10 αποτελέσματα
    })

# Εμφάνιση των αποτελεσμάτων για κάθε ερώτημα
for vsm_result in vsm_results:
    print(f"\nQuery ID: {vsm_result['Query ID']}")  # Εμφάνιση ταυτότητας ερωτήματος
    print(vsm_result['Top 10 Results'])  # Εμφάνιση κορυφαίων 10 αποτελεσμάτων
```

Έπειτα θα προχωρήσουμε στην αξιολόγηση της απόδοσης για τον αλγόριθμο ανάκτησης Vector Space Model (VSM) χρησιμοποιώντας τις εξής μετρικές: 
- Ακρίβεια (Precision)
- Ανάκληση (Recall)
- F1-score 
- Μέση Ακρίβεια Ερωτήματος (AP)
- Μέση Ακρίβεια Ερωτημάτων (MAP)

Μεθοδοι Αξιολόγησης:
- Υλοποιούμε τη συνάρτηση evaluate_vsm_ranking,  με την οποία πραγματοποιούμε evaluation για κάθε ερώτημα για τον αλγόριθμο ανάκτησης Vector Space Model (VSM) και υπολογίζονται η μετρικές precision, recall, F1-score και AP.
- Υλοποιούμε την συνάρτηση evaluate_all_vsm_queries, με την οποία πραγματοποιούμε evaluation για όλα τα ερωτήματα για τον αλγόριθμο ανάκτησης Vector Space Model (VSM) υπολογίζονται οι μέσες τιμές για precision, recall, F1-score καθώς επίσης υπολογίζεται η μέση ακρίβεια (MAP). Η MAP ορίζεται ως η μέση τιμή της AP κάθε ερωτήματος.

```python
from sklearn.metrics import precision_score, recall_score, f1_score  # Εισαγωγή μετρικών αξιολόγησης από το scikit-learn
import numpy as np  # Εισαγωγή της βιβλιοθήκης numpy για υπολογισμούς

def evaluate_vsm_ranking(query_id, retrieved_docs, relevance):
    """Αξιολόγηση ακρίβειας, ανάκλησης, F1-score και μέσης ακρίβειας (AP) για τα αποτελέσματα που ανακτήθηκαν."""
    relevance = {str(k): [str(doc) for doc in v] for k, v in relevance.items()}  # Μετατροπή των IDs σε strings
    query_id = str(query_id)  # Μετατροπή του Query ID σε string
    relevant_docs = set(relevance.get(query_id, []))  # Ανάκτηση των σχετικών εγγράφων για το συγκεκριμένο ερώτημα
    retrieved_relevant = [doc for doc in retrieved_docs if doc in relevant_docs]  # Ανάκτηση των σχετικών εγγράφων που βρέθηκαν

    # Υπολογισμός μετρικών
    precision = len(retrieved_relevant) / len(retrieved_docs) if retrieved_docs else 0  # Ακρίβεια
    recall = len(retrieved_relevant) / len(relevant_docs) if relevant_docs else 0  # Ανάκληση
    f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0  # F1-Score

    # Υπολογισμός Μέσης Ακρίβειας (AP)
    ap = 0
    if relevant_docs:
        relevant_hits = 0
        for i, doc in enumerate(retrieved_docs, start=1):
            if doc in relevant_docs:
                relevant_hits += 1
                ap += relevant_hits / i  # Precision για κάθε σχετικό αποτέλεσμα
        ap /= len(relevant_docs)

    return precision, recall, f1, ap

def evaluate_all_vsm_queries(vsm_results, relevance):
    """Αξιολόγηση όλων των ερωτημάτων με χρήση του VSM και υπολογισμός της Μέσης Ακρίβειας (MAP)."""
    overall_precision, overall_recall, overall_f1, average_precisions = [], [], [], []  # Λίστες για αποθήκευση μετρικών

    for result in vsm_results:
        query_id = result['Query ID']  # Ταυτότητα ερωτήματος
        retrieved_docs = result['Top 10 Results']['Document ID'].tolist()  # Ανάκτηση όλων των εγγράφων που βρέθηκαν

        # Αξιολόγηση της κατάταξης για το συγκεκριμένο ερώτημα
        precision, recall, f1, ap = evaluate_vsm_ranking(query_id, retrieved_docs, relevance)

        # Εμφάνιση αποτελεσμάτων για το ερώτημα
        print(f"\nQuery ID: {query_id}")
        print(f"Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}, Average Precision: {ap:.4f}")

        # Προσθήκη μετρικών στις συνολικές λίστες
        overall_precision.append(precision)
        overall_recall.append(recall)
        overall_f1.append(f1)
        average_precisions.append(ap)

    # Υπολογισμός μέσων μετρικών για όλα τα ερωτήματα
    mean_precision = np.mean(overall_precision)
    mean_recall = np.mean(overall_recall)
    mean_f1 = np.mean(overall_f1)
    map_score = np.mean(average_precisions)

    # Εμφάνιση συνολικών αποτελεσμάτων
    print("\n=== Overall Evaluation for VSM ===")
    print(f"Mean Precision: {mean_precision:.4f}")
    print(f"Mean Recall: {mean_recall:.4f}")
    print(f"Mean F1-Score: {mean_f1:.4f}")
    print(f"Mean Average Precision (MAP): {map_score:.4f}")

# Αξιολόγηση των αποτελεσμάτων του VSM
evaluate_all_vsm_queries(vsm_results, relevance)
```

```python
import numpy as np  # Εισαγωγή της βιβλιοθήκης numpy για αριθμητικούς υπολογισμούς

# Υπολογισμός αξιολόγησης BM25 για ένα μόνο ερώτημα
def evaluate_bm25_ranking(query_id, retrieved_docs, relevance):
    """Υπολογίζει την ακρίβεια (precision), την ανάκληση (recall), το F1-score και το AP για την ανάκτηση BM25."""
    query_id = int(query_id)  # Μετατροπή του ID του ερωτήματος σε ακέραιο
    relevant_docs = set(relevance.get(query_id, []))  # Σχετικά έγγραφα ως ακέραιοι
    retrieved_docs = list(map(int, retrieved_docs))  # Εγγραφές που ανακτήθηκαν ως ακέραιοι
    retrieved_relevant = [doc for doc in retrieved_docs if doc in relevant_docs]  # Εύρεση σχετικών εγγράφων

    # Υπολογισμός ακρίβειας, ανάκλησης, F1-Score
    precision = len(retrieved_relevant) / len(retrieved_docs) if retrieved_docs else 0
    recall = len(retrieved_relevant) / len(relevant_docs) if relevant_docs else 0
    f1 = (2 * precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    # Υπολογισμός μέσης ακρίβειας (AP)
    ap = 0
    if relevant_docs:
        relevant_hits = 0
        for i, doc in enumerate(retrieved_docs, start=1):  # Επανάληψη στα έγγραφα που ανακτήθηκαν
            if doc in relevant_docs:  # Αν το έγγραφο είναι σχετικό
                relevant_hits += 1
                ap += relevant_hits / i  # Υπολογισμός ακρίβειας στο συγκεκριμένο σημείο
        ap /= len(relevant_docs)  # Κανονικοποίηση βάσει των σχετικών εγγράφων

    return precision, recall, f1, ap  # Επιστροφή των μετρικών

# Υπολογισμός αξιολόγησης BM25 για όλα τα ερωτήματα
def evaluate_all_bm25_queries(bm25_results, relevance):
    """Υπολογίζει την ακρίβεια, την ανάκληση, το F1-Score και τη μέση ακρίβεια (MAP) για όλα τα ερωτήματα BM25."""
    overall_precision, overall_recall, overall_f1, average_precisions = [], [], [], []

    for result in bm25_results:  # Επανάληψη στα αποτελέσματα BM25
        query_id = result['Query ID']  # ID του ερωτήματος
        retrieved_docs = result['Top 10 Results']['Document ID'].tolist()  # Ανάκτηση των αναγνωριστικών εγγράφων

        # Υπολογισμός μετρικών για το συγκεκριμένο ερώτημα
        precision, recall, f1, ap = evaluate_bm25_ranking(query_id, retrieved_docs, relevance)

        # Εκτύπωση μετρικών για κάθε ερώτημα
        print(f"\nQuery ID: {query_id}")
        print(f"Precision: {precision:.4f}, Recall: {recall:.4f}, F1-Score: {f1:.4f}, Average Precision: {ap:.4f}")

        # Αποθήκευση των μετρικών
        overall_precision.append(precision)
        overall_recall.append(recall)
        overall_f1.append(f1)
        average_precisions.append(ap)

    # Υπολογισμός συνολικών μετρικών
    mean_precision = np.mean(overall_precision)
    mean_recall = np.mean(overall_recall)
    mean_f1 = np.mean(overall_f1)
    map_score = np.mean(average_precisions)

    # Εκτύπωση συνολικών αποτελεσμάτων
    print("\n=== Overall Evaluation for BM25 ===")
    print(f"Mean Precision: {mean_precision:.4f}")
    print(f"Mean Recall: {mean_recall:.4f}")
    print(f"Mean F1-Score: {mean_f1:.4f}")
    print(f"Mean Average Precision (MAP): {map_score:.4f}")

# Μετατροπή των κλειδιών και τιμών του λεξικού `relevance` σε ακέραιους
relevance = {int(k): list(map(int, v)) for k, v in relevance.items()}

# Αξιολόγηση των αποτελεσμάτων BM25
evaluate_all_bm25_queries(bm25_results, relevance)
```

Στη συνέχεια επιχειρούμε να βελτιώσουμε τα αποτελέσματα της αξιολόγησης των αλγορίθμων επειχειρώντας επιπρόσθετες τεχνικές προεπεξεργασίας των εγγράφων και των ερωτημάτων,εκτός του tokenization, όπως lemmatization, αφαίρεση των stopwords και αφαίρεση των ειδικών χαρακτήρων.

```python
import json
import nltk
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

# Κατέβασμα των απαραίτητων πόρων του NLTK
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

# Αρχικοποίηση του lemmatizer και των stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Βοηθητική συνάρτηση για καθαρισμό και προεπεξεργασία του κειμένου
def get_words(text):
    # Tokenization: Μετατροπή του κειμένου σε πεζά και διαχωρισμός σε λέξεις
    tokens = word_tokenize(text.lower())
    
    # Αφαίρεση ειδικών χαρακτήρων και αριθμών (διατηρούνται μόνο οι αλφαβητικές λέξεις)
    tokens = [token for token in tokens if token.isalpha()]
    
    # Λημματοποίηση: Μετατροπή των λέξεων στη βασική τους μορφή και αφαίρεση των stopwords
    cleaned_tokens = [
        lemmatizer.lemmatize(token) for token in tokens if token not in stop_words
    ]
    
    return cleaned_tokens

# Επεξεργασία των εγγράφων και των ερωτημάτων
doc_words = {}
qry_words = {}

for doc_id in documents.keys():
    doc_words[doc_id] = get_words(documents.get(doc_id))

for qry_id in queries.keys():
    qry_words[qry_id] = get_words(queries.get(qry_id))

# Debugging: Εμφάνιση των πρώτων λέξεων των εγγράφων και των ερωτημάτων
print("\n=== Document Word Lists ===")
print(f"Total number of documents: {len(doc_words)}")
print(f"Words in Document ID 1: {doc_words.get(1)[:20]}...")  # Εμφάνιση αποσπάσματος από τις λέξεις του πρώτου εγγράφου
print(f"Total words in Document ID 1: {len(doc_words.get(1))}")

print("\n=== Query Word Lists ===")
print(f"Total number of queries: {len(qry_words)}")
print(f"Words in Query ID 1: {qry_words.get(1)[:20]}...")  #  Εμφάνιση αποσπάσματος από τις λέξεις του πρώτου ερωτήματος
print(f"Total words in Query ID 1: {len(qry_words.get(1))}")
print(f"Words in Query ID 2: {qry_words.get(2)[:20]}...")  # Εμφάνιση αποσπάσματος από τις λέξεις του δεύτερου ερωτήματος
print(f"Total words in Query ID 2: {len(qry_words.get(2))}")

# Αποθήκευση των επεξεργασμένων λιστών λέξεων σε αρχεία JSON
doc_path = r'C:\Users\07dig\OneDrive\tokenized_documents_improved.json'
qry_path = r'C:\Users\07dig\OneDrive\tokenized_queries_improved.json'

with open(doc_path, 'w') as doc_file:
    json.dump(doc_words, doc_file, indent=4)

with open(qry_path, 'w') as qry_file:
    json.dump(qry_words, qry_file, indent=4)
```
